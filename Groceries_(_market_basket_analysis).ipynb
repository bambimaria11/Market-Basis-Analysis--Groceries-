```python
basket=df.groupby(['Member_number','itemDescription'])['itemDescription'].count().unstack().reset_index().fillna(0).set_index('Member_number')
basket
```

```python
basket ['waffles'].value_counts()
```

```python
def encode(x):
  if x>0:
    return 1
  if x <=0:
    return 0

encoded_basket = basket.applymap(encode)
```

```python
encoded_basket ['waffles'].value_counts()
```

```python
!pip install mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
apr=apriori(encoded_basket,min_support=0.07,use_colnames=True)
```

```python
df_rules = association_rules(apr,metric='lift',min_threshold=1)
df_rules
```

```python
def get_recommendations(product,rules,num_recommendations=5):
    recommendations =[]
    for idx,row in rules.iterrows():
       if product in row ['antecedents']:
        recommended_items = next(iter(row['consequents']))
        recommendations.append((recommended_items,row['lift']))
    recommendations.sort(key=lambda x :x[1],reverse=True)
    return recommendations[:num_recommendations]
```

```python
product_to_recommend = 'yogurt'
recommendations = get_recommendations(product_to_recommend,df_rules)
print(f"recommendations for ' {product_to_recommend}':")
for items,lift in recommendations:
   print(f"{items}(lift:{lift:.2f})")
```
